```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from sklearn.manifold import TSNE
#%matplotlib tk
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
model = TSNE()

x = x_train[0:10000,:]
y = y_train[0:10000]
```

```python
tsne_data = model.fit_transform(x.reshape(x.shape[0], 28*28))
```

```python
#for i in range(10):
#    plt.scatter(tsne_data[:,0][y==i],tsne_data[:,1][y==i], label=i)
```

```python
for i in range(10):
    plt.scatter(tsne_data[:,0][y==i],tsne_data[:,1][y==i], label=i, alpha=0.5)
plt.legend(loc="upper left", bbox_to_anchor=(1.01, 1))
plt.show()
```
